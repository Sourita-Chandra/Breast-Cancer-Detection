## Importing Libraries

```python
import numpy as np
import pandas as pd
```

```python

df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.info()
```

# Data Cleaning

```python
## removing the id column
df.drop(df.columns[[-1,0]], axis=1, inplace=True) 
```

```python
df
```

```python
df['diagnosis'].value_counts()
## B : Benign -- noncancer
## M : Malignant -- cancer
```

# Feature Selection

```python

from sklearn.model_selection import train_test_split
```

```python
## creating a dictionary where M=1 , B=0
diag_map = {'M' : 1, 'B' : 0}
```

```python
df['diagnosis'] = df['diagnosis'].map(diag_map)
```

```python
df
```

```python
y = df[['diagnosis']]
```

```python
## randomly selected 5 features to train the model
##x = df[['radius_mean','perimeter_mean','area_mean','concavity_mean','concave points_mean']]
## here the accuracy was 93%
```

```python
##we again create a new df where all columns are present
df.drop(df.columns[[-1,0]], axis=1, inplace=True) 
x = df
```

```python
df
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=44)
```

## Model - KNN

```python

from sklearn.neighbors import KNeighborsClassifier
```

```python

knn=KNeighborsClassifier()
```

```python

knn.fit(x_train,y_train)
```

```python

knn_y_pred = knn.predict(x_test)
```

```python

from sklearn.metrics import accuracy_score
```

```python

accuracy_score(knn_y_pred, y_test)
```

## Model - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state = 10)
```

```python
lr.fit(x_train,y_train)
```

```python
lr_y_pred=lr.predict(x_test)
```

```python
accuracy_score(lr_y_pred, y_test)
```

## Model - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
# Initialize and train the model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
```

```python
# Make predictions
rf_y_pred = rf.predict(x_test)
```

```python
# Evaluate the model
print("Random Forest Classification Report: \n",classification_report(y_test, rf_y_pred))
print ("Accuracy Score:", accuracy_score(y_test, rf_y_pred))
```

## Model - Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# Initialize and train the Gradient Boosting model
gb = GradientBoostingClassifier(n_estimators=100, random_state=2)
gb.fit(x_train, y_train)
```

```python
# Make predictions
gb_y_pred = gb.predict(x_test)
```

```python
# Evaluate the model
print("Gradient Boosting Classification Report: \n",classification_report(y_test, gb_y_pred))
print("Accuracy Score:", accuracy_score(y_test, gb_y_pred))
```

## Model - XGB Classifier

```python
import xgboost as xgb
```

```python
# Initialize and train the XBoost model
xgb = xgb. XGBClassifier(n_estimators=100, random_state=42, use_label_encoder=False)
xgb.fit(x_train, y_train)
```

```python
# Make predictions
xgb_y_pred = xgb. predict(x_test)
```

```python
# Evaluate the model
print("XGBoost Classification Report:\n", classification_report(y_test, xgb_y_pred))
print("Accuracy Score:", accuracy_score(y_test, xgb_y_pred))
```

```python
from sklearn.metrics import accuracy_score

# Assuming the models are already trained and stored in the variables:
# knn, lr, rf, gb, xgb, gnb

# Dictionary to store model names and their accuracies
model_accuracies = {}

# Calculate and store accuracy for each model
model_accuracies['KNN'] = accuracy_score(y_test, knn.predict(x_test))
model_accuracies['Logistic Regression'] = accuracy_score(y_test, lr.predict(x_test))
model_accuracies['Random Forest'] = accuracy_score(y_test, rf.predict(x_test))
model_accuracies['Gradient Boosting'] = accuracy_score(y_test, gb.predict(x_test))
model_accuracies['XGBoost'] = accuracy_score(y_test, xgb.predict(x_test))
model_accuracies['Naive Bayes'] = accuracy_score(y_test, gnb.predict(x_test))

# Sort the models by accuracy in descending order
sorted_accuracies = dict(sorted(model_accuracies.items(), key=lambda item: item[1], reverse=True))

# Print heading and sorted accuracies
print("Sorted Model Accuracy Scores:")
for model_name, accuracy in sorted_accuracies.items():
    print(f"{model_name}: {accuracy:.2f}")
```
